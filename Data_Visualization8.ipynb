```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(11)
y=x**2
print(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y,'r-*')
```

```python
plt.plot(x,y,'r-*')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
plt.show()
```

```python
plt.plot(x,y,'r-*')
plt.xticks(range(0,8))
plt.yticks(range(0,101,10))
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
plt.show()
```

```python
plt.subplot(3,3,1)
plt.plot(x,y,'r-*')

plt.subplot(2,3,2)
plt.plot(x,y,'g-*')

plt.subplot(1,3,3)
plt.plot(x,y,'b-*')
```

```python
fig=plt.figure(figsize=(10,10))
axes=fig.add_axes([0.1,0.2,0.8,0.3])
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
fig.savefig('graph.pdf',dpi=200)
```

```python


import numpy as np
import pandas as pd

np.random.seed(42)

districts = ['Downtown', 'East Side', 'West End', 'Northville', 'Southtown']
complaint_types = ['Power Cut', 'Traffic', 'Traffick', 'Noise', 'Garbage',
                   'Water Leak', 'Power Cut', 'Ice Cream']
ice_cream_flavors = ['Vanilla', 'Chocolate', 'Strawberry', 'Mint', 'Mango',
                     'Lemon', 'Pistachio', 'Cookie Dough']

n_rows = 500

data = {
    'district': np.random.choice(districts, size=n_rows),
    'complaint_type': np.random.choice(complaint_types, size=n_rows),
    'temperature': np.random.normal(loc=32, scale=5, size=n_rows),
    'electricity_usage': np.random.normal(loc=250, scale=50, size=n_rows),
    'num_ice_cream_carts': np.random.choice([0, 2, 5, 10, 10, 10], size=n_rows),
    'survey_rating': np.random.choice([1, 2, 3, 4, 5, np.nan], size=n_rows),
}

df = pd.DataFrame(data)
df.head()
```

```python
df.head(10)
```

```python
df.loc[5:10,'complaint_type']=np.nan
```

```python
df.head(10)
```

```python
df=pd.concat([df,df.iloc[0:5]])
```

```python
import seaborn as sns
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.histplot(df['temperature'],bins=20,kde=True)
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='district',y='electricity_usage',data=df)
plt.title('Electricity Usage by District')
plt.xlabel('District')
plt.ylabel('Electricity Usage')
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='plasma')
plt.title('Correlation Heatmap')
plt.show()
```
